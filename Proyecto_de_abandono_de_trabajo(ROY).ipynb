# Pruebas de Hipótesis

```python
# Convertir la columna 'satisfaccion_trabajo' a numérica (si es necesario)
df_split['satisfaccion_trabajo'] = pd.to_numeric(df_split['satisfaccion_trabajo'], errors='coerce')

# Eliminar filas con valores nulos en 'abandono' o 'satisfaccion_trabajo'
df_split = df_split.dropna(subset=['abandono', 'satisfaccion_trabajo'])

# Convertir la columna 'abandono' a binaria
df_split['abandono'] = df_split['abandono'].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)

# Dividir los datos en dos grupos: abandono y no abandono
grupo_abandono = df_split[df_split['abandono'] == 1]['satisfaccion_trabajo']
grupo_no_abandono = df_split[df_split['abandono'] == 0]['satisfaccion_trabajo']

# Realizar el test t para dos muestras independientes
t_stat, p_value = ttest_ind(grupo_abandono, grupo_no_abandono)
print(f'Estadístico t: {t_stat}, Valor p: {p_value}')

# Prueba t para dos muestras independientes (ejemplo con 'salario_mes' y 'abandono')
grupo_abandono = df_split[df_split['abandono'] == 'yes']['salario_mes']
grupo_no_abandono = df_split[df_split['abandono'] == 'no']['salario_mes']

t_stat, p_value = ttest_ind(grupo_abandono.dropna(), grupo_no_abandono.dropna())
print(f'Estadístico t: {t_stat}, Valor p: {p_value}')

# Visualización con gráficos
plt.figure(figsize=(12, 6))

# Gráfico de violín
plt.subplot(1, 2, 1)
sns.violinplot(x='abandono', y='satisfaccion_trabajo', data=df_split, palette='muted')
plt.title('Distribución de la Satisfacción en el Trabajo por Abandono')
plt.xlabel('Abandono')
plt.ylabel('Satisfacción en el Trabajo')
plt.xticks([0, 1], ['No', 'Yes'])

# Gráfico de cajas
plt.subplot(1, 2, 2)
sns.boxplot(x='abandono', y='satisfaccion_trabajo', data=df_split, palette='muted')
plt.title('Caja de la Satisfacción en el Trabajo por Abandono')
plt.xlabel('Abandono')
plt.ylabel('Satisfacción en el Trabajo')
plt.xticks([0, 1], ['No', 'Yes'])

```

# Árboles de Decision:
son modelos de predicción utilizados tanto para clasificación como para regresión. Un árbol de decisión divide los datos en subconjuntos basándose en las características de los datos, creando un modelo en forma de estructura de árbol.

```python
# preparar los datos
X = df.drop(columns=['abandono'])
y = df['abandono']

# Convert categorical features to numerical using one-hot encoding
X = pd.get_dummies(X)  # Convert categorical columns to numerical

# Dividir datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

# Entrenar modelo
dt = DecisionTreeClassifier(max_depth=7, random_state=7) # Remove n_estimators
dt.fit(X_train, y_train)

# Predecir
y_pred = dt.predict(X_test)

# Evaluar
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')


# reporte de clasificacion
print("classification Report:")
print(classification_report(y_test, y_pred))


# Visualizar arbol
plt.figure(figsize=(10, 8))
plot_tree(dt, feature_names=X.columns, filled=True)
plt.xlabel('Predicted')
plt.ylabel('true')
plt.title ('Confusion Matrix')
plt.show()
```

# Análisis de las Razones del Abandono Empleados
   # Edad y Experiencia:

- Edad:
 Los empleados más jóvenes y aquellos cercanos a la jubilación muestran una mayor probabilidad de abandonar.
- Años en la Compañía: Aquellos con menor tiempo en la compañía tienden a abandonar más, indicando que los primeros años son cruciales para la retención.
Satisfacción:

- Satisfacción en el Entorno: La insatisfacción con el entorno laboral, ya sea por condiciones físicas o ambiente de trabajo, es un factor importante.
- Satisfacción en el Trabajo: La insatisfacción con las tareas y responsabilidades diarias es un predictor significativo del abandono.
# Balance Trabajo y Vida:

- Conciliación: La dificultad para equilibrar el trabajo con la vida personal, especialmente en términos de horarios flexibles y trabajo remoto, contribuye al abandono.
- Horas Extra: Los empleados que trabajan muchas horas extra tienden a abandonar más, lo cual puede ser un signo de sobrecarga laboral.
# Progresión y Desarrollo:

- Promociones y Desarrollo Profesional: La falta de oportunidades de promoción y desarrollo profesional puede desmotivar a los empleados.
- Formaciones Recientes: Aquellos que no reciben capacitación o formación regular también muestran mayores tasas de abandono.
- Condiciones Laborales y Compensación:

- Salario y Beneficios: La percepción de una compensación justa es crucial. Los empleados que sienten que no están bien remunerados en comparación con el mercado tienden a abandonar.
- Evaluación de Desempeño: La falta de reconocimiento y evaluación justa del desempeño puede desmotivar a los empleados.
- Recomendaciones para Mejorar la Retención de Empleados
- Mejorar la Satisfacción en el Entorno y en el Trabajo:

- Realizar encuestas regulares de satisfacción para identificar y abordar problemas en el ambiente laboral.
-Fomentar un entorno de trabajo positivo y colaborativo, incluyendo espacios adecuados y cómodos.
-Promover un Balance Trabajo-Vida Saludable:

- Implementar políticas de trabajo flexible y remoto.
- Regular las horas extra y asegurarse de que no se conviertan en una norma.
- Fomentar el Desarrollo Profesional:

- Crear y comunicar claras trayectorias de carrera y oportunidades de promoción.
- Ofrecer programas de capacitación y desarrollo continuos.
- Asegurar una Compensación Competitiva:

- Revisar y ajustar los paquetes salariales y de beneficios regularmente para asegurarse de que son competitivos en el mercado.
-Implementar programas de reconocimiento y recompensas basados en el desempeño.
- Evaluar y Mejorar la Gestión:

- Capacitar a los managers para que sean líderes efectivos y puedan apoyar mejor a sus equipos.
- Promover una comunicación abierta y feedback continuo entre empleados y gestores.
# Conclusión
El análisis de los datos sugiere que la satisfacción en el entorno y en el trabajo, el balance trabajo-vida, el desarrollo profesional, y las condiciones laborales y de compensación son factores críticos que influyen en la decisión de los empleados de abandonar la empresa. Implementar mejoras en estas áreas puede contribuir significativamente a la retención de empleados, aumentando la satisfacción y el compromiso de los mismos. Al enfocarse en estas áreas clave, las empresas pueden crear un entorno más atractivo y retentivo para sus empleados, reduciendo así la tasa de abandono.
